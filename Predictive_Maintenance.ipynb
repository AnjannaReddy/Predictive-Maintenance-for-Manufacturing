```python

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import (
    roc_auc_score, precision_score, recall_score, f1_score,
    confusion_matrix, accuracy_score, classification_report
)
from sklearn.preprocessing import StandardScaler
import joblib
import xgboost as xgb
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Masking
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# -----------------------------
# 1) Synthetic data generation
# -----------------------------
def generate_synthetic_data(n_machines=100, max_cycles=300, seed=42):
    """
    Generate synthetic sensor time series per machine.
    Each machine operates until failure; failure cycle is randomized,
    sensors degrade as cycle approaches failure.

    Returns: DataFrame with columns:
        machine_id, cycle, vibration, temperature, pressure, failure (0/1)
    """
    np.random.seed(seed)
    rows = []
    for m in range(n_machines):
        # Random failure cycle between 150 and max_cycles
        failure_cycle = np.random.randint(120, max_cycles + 1)
        for cycle in range(1, failure_cycle + 1):
            # base signals
            base_vib = 0.5 + 0.001 * cycle  # slight trend
            base_temp = 30 + 0.02 * cycle
            base_pres = 5 + 0.005 * cycle

            # degradation component grows near failure
            progress = cycle / failure_cycle
            degr_factor = (progress ** 6)  # slow then steep
            vibration = base_vib + 3.0 * degr_factor + np.random.normal(0, 0.05)
            temperature = base_temp + 10.0 * degr_factor + np.random.normal(0, 0.2)
            pressure = base_pres + 1.5 * degr_factor + np.random.normal(0, 0.02)

            # failure flag only at last cycle
            failure = 1 if cycle == failure_cycle else 0

            rows.append({
                "machine_id": f"m_{m}",
                "cycle": cycle,
                "vibration": vibration,
                "temperature": temperature,
                "pressure": pressure,
                "failure": failure,
                "failure_cycle": failure_cycle
            })
    df = pd.DataFrame(rows)
    return df

# -----------------------------
# 2) Labeling for classification
# -----------------------------
def add_failure_within_k_label(df, k=10):
    """
    Add label 'will_fail_within_k' which is 1 if the machine will fail within next k cycles.
    This transforms the problem into supervised classification.
    """
    df = df.copy()
    # for each row, check if failure_cycle - cycle <= k
    df['will_fail_within_k'] = (df['failure_cycle'] - df['cycle'] <= k).astype(int)
    return df

# -----------------------------
# 3) Feature engineering
# -----------------------------
def feature_engineering(df):
    """
    Creates rolling statistics, lags, and FFT energy for each sensor per machine.
    """
    df = df.copy()
    df = df.sort_values(['machine_id', 'cycle'])
    # Rolling windows
    window_sizes = [3, 5, 10]
    sensors = ['vibration', 'temperature', 'pressure']
    for w in window_sizes:
        for s in sensors:
            df[f'{s}_roll_mean_{w}'] = df.groupby('machine_id')[s].transform(lambda x: x.rolling(window=w, min_periods=1).mean())
            df[f'{s}_roll_std_{w}'] = df.groupby('machine_id')[s].transform(lambda x: x.rolling(window=w, min_periods=1).std().fillna(0))
    # Lags
    for lag in [1, 2, 3]:
        for s in sensors:
            df[f'{s}_lag_{lag}'] = df.groupby('machine_id')[s].shift(lag).fillna(method='bfill')
    # Simple FFT energy across a small window (using window 8)
    def fft_energy(x):
        # x is a pandas series
        import numpy as _np
        arr = _np.asarray(x)
        # zero pad if small
        if len(arr) < 2:
            return 0.0
        fft = np.fft.rfft(arr - arr.mean())
        # energy in frequency domain
        energy = np.sum(np.abs(fft) ** 2)
        return energy
    df['vibration_fft_energy'] = df.groupby('machine_id')['vibration'].transform(lambda x: x.rolling(window=8, min_periods=2).apply(lambda y: fft_energy(y), raw=False))
    df['temperature_fft_energy'] = df.groupby('machine_id')['temperature'].transform(lambda x: x.rolling(window=8, min_periods=2).apply(lambda y: fft_energy(y), raw=False))
    df['pressure_fft_energy'] = df.groupby('machine_id')['pressure'].transform(lambda x: x.rolling(window=8, min_periods=2).apply(lambda y: fft_energy(y), raw=False))
    # Fill NaNs produced by FFT windows
    df.fillna(method='bfill', inplace=True)
    df.fillna(0, inplace=True)
    return df

# -----------------------------
# 4) Prepare tabular dataset
# -----------------------------
def prepare_tabular_dataset(df, label_col='will_fail_within_k'):
    """
    Select features and split into X, y.
    """
    df = df.copy()
    exclude_cols = ['machine_id', 'failure_cycle', 'failure', 'cycle']
    # define features as all numeric columns except excluded and labels
    features = [c for c in df.columns if c not in exclude_cols + [label_col]]
    X = df[features]
    y = df[label_col].astype(int)
    return X, y, features

# -----------------------------
# 5) Train tree-based models
# -----------------------------
def train_tree_models(X_train, y_train, X_val, y_val):
    # scale features for tree models not strictly necessary but helps xgboost sometimes
    scaler = StandardScaler()
    X_train_s = scaler.fit_transform(X_train)
    X_val_s = scaler.transform(X_val)

    rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
    rf.fit(X_train_s, y_train)

    xgb_clf = xgb.XGBClassifier(n_estimators=200, use_label_encoder=False, eval_metric='logloss', random_state=42, n_jobs=-1)
    xgb_clf.fit(X_train_s, y_train)

    # validations
    preds_rf = rf.predict(X_val_s)
    preds_xgb = xgb_clf.predict(X_val_s)
    proba_rf = rf.predict_proba(X_val_s)[:, 1]
    proba_xgb = xgb_clf.predict_proba(X_val_s)[:, 1]

    return {
        'scaler': scaler,
        'rf': rf,
        'xgb': xgb_clf,
        'rf_preds': preds_rf,
        'xgb_preds': preds_xgb,
        'rf_proba': proba_rf,
        'xgb_proba': proba_xgb
    }

# -----------------------------
# 6) LSTM sequence model
# -----------------------------
def create_sequence_windows(df, sensors=['vibration','temperature','pressure'], window_size=30, step=1, label_col='will_fail_within_k'):
    """
    Create sliding windows per machine for LSTM.
    Returns X_seq: (n_windows, window_size, n_features), y_seq: (n_windows,)
    The label is 1 if within the last row of the window label_col ==1 (you can adapt)
    """
    Xs = []
    ys = []
    groups = df.groupby('machine_id')
    for mid, g in groups:
        g = g.sort_values('cycle')
        arr = g[sensors].values
        labels = g[label_col].values
        n = len(g)
        if n < window_size:
            continue
        for start in range(0, n - window_size + 1, step):
            end = start + window_size
            Xs.append(arr[start:end, :])
            # label: will it fail within k from the last cycle of the window? use label at last pos
            ys.append(labels[end - 1])
    Xs = np.asarray(Xs)
    ys = np.asarray(ys)
    return Xs, ys

def build_lstm_model(window_size, n_features):
    model = Sequential([
        Masking(mask_value=0., input_shape=(window_size, n_features)),
        LSTM(64, return_sequences=False),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# -----------------------------
# 7) Utilities: evaluation
# -----------------------------
def evaluate_classification(y_true, y_pred, y_proba=None, name="model"):
    print(f"\n=== Evaluation: {name} ===")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, zero_division=0))
    print("Recall:", recall_score(y_true, y_pred, zero_division=0))
    print("F1:", f1_score(y_true, y_pred, zero_division=0))
    if y_proba is not None:
        try:
            print("ROC AUC:", roc_auc_score(y_true, y_proba))
        except:
            pass
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("Classification Report:\n", classification_report(y_true, y_pred, zero_division=0))

# -----------------------------
# 8) Main pipeline
# -----------------------------
def main():
    # 1. Generate synthetic data
    print("Generating synthetic data...")
    df = generate_synthetic_data(n_machines=150, max_cycles=400, seed=123)
    print("Data shape:", df.shape)
    # optional: replace above with real CSV load:
    # df = pd.read_csv("your_sensor_data.csv")

    # 2. Label: will fail within next k cycles? (k=10)
    k = 10
    df = add_failure_within_k_label(df, k=k)

    # 3. Feature engineering
    print("Feature engineering...")
    df_feat = feature_engineering(df)

    # 4. Prepare tabular dataset
    X, y, feature_cols = prepare_tabular_dataset(df_feat, label_col='will_fail_within_k')
    print("Feature columns:", len(feature_cols))

    # 5. Split into train/val/test by machine to prevent leakage
    machine_ids = df_feat['machine_id'].unique()
    train_m, test_m = train_test_split(machine_ids, test_size=0.3, random_state=42)
    # further split train into train/val
    train_m, val_m = train_test_split(train_m, test_size=0.15, random_state=42)

    train_idx = df_feat['machine_id'].isin(train_m)
    val_idx = df_feat['machine_id'].isin(val_m)
    test_idx = df_feat['machine_id'].isin(test_m)

    X_train, y_train = X[train_idx], y[train_idx]
    X_val, y_val = X[val_idx], y[val_idx]
    X_test, y_test = X[test_idx], y[test_idx]

    print("Train/Val/Test sizes:", X_train.shape[0], X_val.shape[0], X_test.shape[0])
    # 6. Train tree models
    print("Training RandomForest and XGBoost...")
    tree_results = train_tree_models(X_train, y_train, X_val, y_val)

    # Evaluate on validation
    evaluate_classification(y_val, tree_results['rf_preds'], tree_results['rf_proba'], name="RandomForest (val)")
    evaluate_classification(y_val, tree_results['xgb_preds'], tree_results['xgb_proba'], name="XGBoost (val)")

    # Save tree models and scaler
    os.makedirs('models', exist_ok=True)
    joblib.dump(tree_results['scaler'], 'models/scaler.joblib')
    joblib.dump(tree_results['rf'], 'models/rf_model.joblib')
    joblib.dump(tree_results['xgb'], 'models/xgb_model.joblib')
    print("Saved tree models in ./models/")

    # Evaluate on test
    scaler = tree_results['scaler']
    X_test_s = scaler.transform(X_test)
    rf_test_preds = tree_results['rf'].predict(X_test_s)
    rf_test_proba = tree_results['rf'].predict_proba(X_test_s)[:, 1]
    evaluate_classification(y_test, rf_test_preds, rf_test_proba, name="RandomForest (test)")

    # 7. LSTM sequence model
    print("Preparing sequence windows for LSTM...")
    sensors = ['vibration', 'temperature', 'pressure']
    window_size = 30
    X_seq, y_seq = create_sequence_windows(df_feat, sensors=sensors, window_size=window_size, step=5, label_col='will_fail_within_k')
    print("Sequence dataset shape:", X_seq.shape, y_seq.shape)

    # Split sequences by machines: simple random split (better to split by machine)
    Xs_train, Xs_test, ys_train, ys_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42, stratify=y_seq)
    # further split validation
    Xs_train, Xs_val, ys_train, ys_val = train_test_split(Xs_train, ys_train, test_size=0.15, random_state=42, stratify=ys_train)

    # Scale features per sensor (fit on training windows)
    nsamples, w, nf = Xs_train.shape
    scaler_seq = StandardScaler()
    Xs_train_reshaped = Xs_train.reshape(-1, nf)
    scaler_seq.fit(Xs_train_reshaped)
    def scale_seq(X):
        s = X.reshape(-1, X.shape[2])
        s = scaler_seq.transform(s)
        return s.reshape(X.shape)
    Xs_train_s = scale_seq(Xs_train)
    Xs_val_s = scale_seq(Xs_val)
    Xs_test_s = scale_seq(Xs_test)

    # build & train model
    print("Building LSTM model...")
    lstm = build_lstm_model(window_size=window_size, n_features=len(sensors))
    lstm.summary()
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True),
        ModelCheckpoint('models/lstm_best.h5', monitor='val_loss', save_best_only=True)
    ]
    history = lstm.fit(
        Xs_train_s, ys_train,
        validation_data=(Xs_val_s, ys_val),
        epochs=40,
        batch_size=64,
        callbacks=callbacks,
        verbose=2
    )

    # Evaluate LSTM
    ys_pred_proba = lstm.predict(Xs_test_s).ravel()
    ys_pred = (ys_pred_proba >= 0.5).astype(int)
    evaluate_classification(ys_test, ys_pred, ys_pred_proba, name="LSTM (test)")
    lstm.save('models/lstm_final.h5')
    joblib.dump(scaler_seq, 'models/scaler_seq.joblib')
    print("Saved LSTM model and sequence scaler in ./models/")

    # Visualization (optional): show one machine's sensors with failure marker
    print("Plotting sample machine signals...")
    sample_mid = df_feat['machine_id'].unique()[0]
    sample = df_feat[df_feat['machine_id'] == sample_mid]
    fig, ax = plt.subplots(3, 1, figsize=(10, 8), sharex=True)
    ax[0].plot(sample['cycle'], sample['vibration'])
    ax[0].set_title(f"{sample_mid} - vibration")
    ax[1].plot(sample['cycle'], sample['temperature'])
    ax[1].set_title("temperature")
    ax[2].plot(sample['cycle'], sample['pressure'])
    ax[2].set_title("pressure")
    plt.tight_layout()
    plt.savefig("sample_signals.png")
    print("Saved sample plot: sample_signals.png")

    print("Pipeline finished.")

if __name__ == "__main__":
    main()
```
